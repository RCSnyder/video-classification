```python
def get_dir_names_all_videos(home_dir):
    """returns the names in a directories subfolders"""
    video_fnames = []
    for path, subdirs, files in os.walk(home_dir):
        for name in files:
            video_fnames.append(str(Path(os.path.join(path, name))))
    return video_fnames
```

```python
def resize_videos(input_dir, resize_x_dim, resize_y_dim):
    """Copies the folder structure of the input_directory to the 
    output directory and resizes the frames within each folder to
    (resize_x_dim X resize_y_dim) in the input directory """
    frames_dirs = get_dir_names_all_videos(input_dir)
    new_frame_dirs = replace_target_directory(frames_dirs, "video_frames")
    
    dim = (resize_x_dim, resize_y_dim)
    
    for frame_dir, new_frame_dir in tqdm_notebook(zip(frames_dirs, new_frame_dirs)):
        img = cv2.imread(frame_dir, cv2.IMREAD_UNCHANGED)

        # resize image
        resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
        
        cv2.imwrite( str(new_frame_dir), resized)
```

```python
def replace_one_target_directory(old_dir, new_dir_name):
    """replaces the 5th subfolder in a directory with a new_dir_name"""
    new_dir = old_dir.split("\\")
    new_dir[5] = new_dir_name
    new_dir = "\\".join(new_dir)
    return new_dir
```

```python
def replace_target_directory(video_names, new_directory_name):
    """replaces the 5th directory with the new directory name"""
    new_dirs = []
    for fname in video_names:
        new_dir = fname.split("\\")
        new_dir[5] = new_directory_name
        new_dir = "\\".join(new_dir)
        new_dirs.append(new_dir)
    return new_dirs
```
